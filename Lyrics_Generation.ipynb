```python
lyrics = pd.read_csv('/content/drive/MyDrive/lyrics/lyrics-data.csv')
artists = pd.read_csv('/content/drive/MyDrive/lyrics/artists-data.csv')
```

```python
lyrics = lyrics[lyrics['language']=='en']
lyrics['language'].value_counts()
```

```python
artists.head()
```

```python
lyrics = lyrics.drop(columns = ['ALink', 'SLink', 'language'])
lyrics.head()
```

```python
lyrics_train_full, lyrics_test = train_test_split(lyrics, train_size = 0.9, random_state = 1)
lyrics_train, lyrics_val = train_test_split(lyrics_train_full, train_size = 0.8, random_state = 1)
```

```python
def build_dataset(df, dest_path):
    f = open(dest_path, 'w')
    data = ''
    lyrics_df = df['Lyric'].tolist()
    for lyric in lyrics_df:
        lyric = str(lyric).strip()
        lyric = re.sub(r"\s", " ", lyric)
        bos_token = '<BOS>'
        eos_token = '<EOS>'
        data += bos_token + ' ' + lyric + ' ' + eos_token + '\n'
        
    f.write(data)
```

```python
build_dataset(lyrics_train, "lyrics_train.txt")
build_dataset(lyrics_val, "lyrics_val.txt")
build_dataset(lyrics_test, "lyrics_test.txt")
```
